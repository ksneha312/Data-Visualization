```python
#Q5:-Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed
np.random.seed(42)

# Number of samples
n_samples = 200

# Define a covariance matrix for 5 features with correlations
cov = np.array([
    [1.0, 0.8, 0.3, 0.0, 0.0],
    [0.8, 1.0, 0.4, 0.0, 0.0],
    [0.3, 0.4, 1.0, 0.5, 0.0],
    [0.0, 0.0, 0.5, 1.0, 0.7],
    [0.0, 0.0, 0.0, 0.7, 1.0]
])

# Mean for each feature
mean = np.zeros(5)

# Generate multivariate normal data
data = np.random.multivariate_normal(mean, cov, size=n_samples)

# Create DataFrame with feature names
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# Calculate correlation matrix
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Q5:-Create a Bokeh heatmap using the provided dataset. data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy = np.meshgrid(x, y)
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.palettes import Viridis256

output_notebook()

# Your data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Bokeh requires the image data as a 2D list inside a list, and coordinates for lower-left corner + width/height
# We'll plot the heatmap as one image spanning from x=0 to 1 and y=0 to 1

# Setup color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

p = figure(title="Heatmap", x_range=(0,1), y_range=(0,1),
           x_axis_location="above", tools="hover", tooltips=[('Value', '@image')],
           width=500, height=500)

# Plot image (note: data_heatmap needs to be flipped vertically for correct orientation)
p.image(image=[data_heatmap[::-1]], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0,0))
p.add_layout(color_bar, 'right')

p.axis.visible = True
p.grid.visible = False

show(p)
```
